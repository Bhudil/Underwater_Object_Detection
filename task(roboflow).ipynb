1. **Augmentations**:
    Here, augmentations like random flip, rotate, and brightness are added to improve model generalization.
    - Random flip: Helps the model recognize objects regardless of orientation.
    - Brightness: Adjusts brightness, contrast, saturation to make model robust to lighting variations.
    - Rotate: Forces model to learn partial object detection at different angles


 2. **Epochs**:
 Chose 50 epochs because it is enough considering resource limitations

 3. **Other Hyperparameters**: Explained below
    - Batch size (16): Balances memory usage and model update frequency for stability.
    - Optimizer (Adam): Adaptive optimizer suitable for faster convergence.
    - Image size (640): Resized to improve detection accuracy with balanced resource usage.

 4. **Output Metrics**:
    Metrics being analyzed include mAP, Precision, and Recall:
    - mAP (mean Average Precision): Measures overall detection accuracy.
    - Precision and Recall: Analyzed to see model balance between true positives and false positives.
